# 6. 데이터 변환(시간데이터 변환)

### 시간 데이터가 object 형태인데, 시간 형태로 바꾸고, 초 단위로 바꿔보자

### 방법1. 사용자 정의 함수 : 시간을 정제하는 함수 만들기

```python
marathon_2017_cleaned['Official Time']
```

### [Pandas] 문자열(str) 처리

```python
marathon_2017_cleaned['Official Time'].str.len()              # 길이 반환  # 총 숫자가 7개라는뜻
```

```python
marathon_2017_cleaned['Official Time'][0]
```

```python
hms = marathon_2017_cleaned['Official Time'].str.split(':', n=2, expand=True)
hms
```

```python
hms[0].astype(int)
```

```python
# 문자열 형태로 되어있는 시간을 int type으로 바꿔준 후 초 단위로 바꿔주는 함수
def to_second(record): # to_second라는 함수를 정의
    hms = record.str.split(':', n=2, expand=True)  # n=2 : 2번 나눠라
    second = hms[0].astype(int)*3600 + hms[1].astype(int)*60 + hms[2].astype(int)
    return second 
```

```python
marathon_2017_cleaned.head()
```

```python
# 위에서 정의한 to_second함수 사용
marathon_2017_cleaned['Official Time Sec1'] = to_second(marathon_2017_cleaned['Official Time'])
```

```python
marathon_2017_cleaned.head()
```

### 방법2. pandas 내장함수 이용 : .to_timedelta() : 시간값을 변환해줌

```python
# timedelta는 시스템에서 날짜시분초를 인식하는 시간 자료형
pd.to_timedelta(marathon_2017_cleaned['Official Time'])[0]
```

```python
import numpy as np
# 초 단위로 바꿔준(astype('m8[s]')) 후 int type으로 바꿔주기(astype(np.int64))
# m8[m] 분 단위로 바꿈
marathon_2017_cleaned['Official Time Sec2'] = pd.to_timedelta(marathon_2017_cleaned['Official Time']).astype('m8[s]').astype(np.int64)
```

```python
marathon_2017_cleaned['Official Time min'] = pd.to_timedelta(marathon_2017_cleaned['Official Time']).astype('m8[m]').astype(np.float64)
marathon_2017_cleaned.head()
```

# 7. 데이터 합치기 : concat

```python
import pandas as pd

marathon_2015 = pd.read_csv('data/marathon_results_2015.csv')
marathon_2016 = pd.read_csv('data/marathon_results_2016.csv')
marathon_2017 = pd.read_csv('data/marathon_results_2017.csv')
```

```python
print(marathon_2015.shape)
print(marathon_2016.shape)
print(marathon_2017.shape)
```

```python
marathon_2015['Year'] = '2015'
marathon_2016['Year'] = '2016'
marathon_2017['Year'] = '2017'
```

```python
# index = 'Offical Time'으로 2015, 2017 데이터 합치기
marathon_2015_2017 = pd.concat([marathon_2015, marathon_2016, marathon_2017], ignore_index=True, sort=False)
```

```python
print(len(marathon_2015_2017.columns))
marathon_2015_2017.columns
```

```python
print(marathon_2015_2017.shape)
marathon_2015_2017.head()
```

### 시간 데이터 변환

```python
import numpy as np

# pd.to_timedelt() 이용하기
# .astype('m8[s]').astype(np.int64) 적용하기 : 초 단위로 바꾼 후, int 형태로 변환
marathon_2015_2017['5K'] = pd.to_timedelta(marathon_2015_2017['5K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['10K'] = pd.to_timedelta(marathon_2015_2017['10K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['15K'] = pd.to_timedelta(marathon_2015_2017['15K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['20K'] = pd.to_timedelta(marathon_2015_2017['20K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['Half'] = pd.to_timedelta(marathon_2015_2017['Half']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['25K'] = pd.to_timedelta(marathon_2015_2017['25K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['30K'] = pd.to_timedelta(marathon_2015_2017['30K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['35K'] = pd.to_timedelta(marathon_2015_2017['35K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['40K'] = pd.to_timedelta(marathon_2015_2017['40K']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['Pace'] = pd.to_timedelta(marathon_2015_2017['Pace']).astype('m8[s]').astype(np.int64)
marathon_2015_2017['Official Time'] = pd.to_timedelta(marathon_2015_2017['Official Time']).astype('m8[s]').astype(np.int64)
```
